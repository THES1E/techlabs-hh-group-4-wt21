```python
# Download data from UK Police (see README) and save the monthly subfolders in the data folder of your local clone of our project.
import glob
import pandas as pd
import matplotlib.pyplot as plt

path =r'../data'
filenames = glob.glob(path + "/*/*.csv")
```

```python
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one dataframe.
df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
# There is one spelling mistake in the data so replace this spelling mistake. 
df['Location'].replace('On or near Shepherds Bush Market (Lu Staion)','On or near Shepherds Bush Market (Lu Station)',inplace=True)
```

```python
# Add a new column called 'Location type' as a copy of the column 'Location' and display the dataframe again.
df['Location type']=df['Location']
df
```

```python
# Change the data type of 'Location type' to string.
df['Location type'] = df['Location type'].astype('string')
```

```python
df.dtypes['Location type']
```

```python
# Define the coordinates that frame the area of Greater London.
lat_max=(df['Latitude']<51.691536)
lat_min=(df['Latitude']>51.291436341041205)
lon_max=(df['Longitude']<0.3337516278548758)
lon_min=(df['Longitude']>-0.5135695653558527)
```

```python
# Filter crimes which are lying in the selected area.
df_filt=df.loc[lat_min & lat_max & lon_min & lon_max]
```

```python
# Display the filtered dataframe.
df_filt
```

```python
# Create a seperate dataframe called 'df_bt' which only contains the crimes that were reported by British Transport Police.
df_bt=df_filt.loc[df_filt['Reported by']=='British Transport Police']
```

```python
df_bt
```

```python
# Clean the dataframe 'df_filt' from the crimes which were reported by British Transport Police to cluster the Location types in both new DataFrames seperately.
df_filt.drop(df_filt.loc[df_filt['Reported by']=='British Transport Police'].index, inplace=True)
```

```python
df_filt
```

```python
# In the dataframe 'df_filt' cluster all locations that happened on any streets in Greater London into the Location type 'On Street'.
df_filt.loc[~df_filt['Location type'].str.contains('On or near Supermarket|On or near Parking Area|On or near Shopping Area|On or near Petrol Station|On or near Sports/Recreation Area|No Location|On or near Nightclub|On or near Pedestrian Subway|On or near Further/Higher Educational Building|On or near Hospital|On or near Police Station|On or near Theatre/Concert Hall|On or near Conference/Exhibition Centre|On or near Park/Open Space|On or near Bus/Coach Station|On or near Meridian Square|On or near Airport/Airfield'), 'Location type'] = 'On Street'
```

```python
# Show all values of Location type and their counts in the dataframe 'df_filt'.
df_filt['Location type'].value_counts()
```

```python
# Replace the Location type 'On or near Pedestrian Subway' with the name 'On or near Underpass'.
df_filt['Location type'].replace('On or near Pedestrian Subway','On or near Underpass',inplace=True)
```

```python
# There are still three street names that include the word hospital that weren't included in the Location type 'On Street' due to that. Replace these three street names with the Location type 'On Street'.
df_filt['Location type'].replace('On or near Hospital Road','On Street',inplace=True)
```

```python
df_filt['Location type'].replace('On or near Hospital Bridge Road','On Street',inplace=True)
```

```python
df_filt['Location type'].replace('On or near Hospital Way','On Street',inplace=True)
```

```python
# Show again all values of Location type and their counts in the dataframe 'df_filt' to see if the clustering is now correct (just Location types and no single street names left).
df_filt['Location type'].value_counts()
```

```python
# In the dataframe 'df_bt' cluster all locations that happened in any London Underground Stations (Lu Station) in Greater London into the Location type 'On or near London Underground Station'.
df_bt.loc[df_bt['Location type'].str.contains('Lu Station'), 'Location type'] = 'On or near London Underground Station'
```

```python
# Show all values of Location type and their counts in the dataframe 'df_bt'.
df_bt['Location type'].value_counts().head(60)
```

```python
# In the dataframe 'df_bt' cluster all locations that happened in any Train Station (Station) or at a Tram Stop in Greater London into the Location type 'On or near Train Station/Tram Stop'.
df_bt.loc[~df_bt['Location type'].str.contains('On or near London Underground Station'), 'Location type'] = 'On or near Train Station/Tram Stop'
```

```python
# Show again all values of Location type and their counts in the dataframe 'df_bt'.
df_bt['Location type'].value_counts()
```

```python
# Concatenate the two dataframes 'df_filt' and 'df_bt' into one dataframe called 'df'.
df = pd.concat([df_filt, df_bt], ignore_index=True, sort=False)
```
